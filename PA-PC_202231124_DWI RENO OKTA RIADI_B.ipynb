```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_images(images, titles, cmap=None):
    plt.figure(figsize=(15, 15))
    for i, (image, title) in enumerate(zip(images, titles)):
        plt.subplot(2, 2, i + 1)
        if cmap:
            plt.imshow(image, cmap=cmap)
        else:
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title(title)
        plt.xticks([]), plt.yticks([])  # Hilangkan sumbu
    plt.show()
```

```python
image_path = 'PA-PC_202231124_DWI RENO OKTA RIADI_B.jpg'  
image = cv2.imread(image_path)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
lower_belimbing = np.array([20, 100, 100])
upper_belimbing = np.array([70, 255, 255])

lower_green = np.array([45, 40, 40])
upper_green = np.array([100, 255, 255])
```

```python
mask_belimbing = cv2.inRange(hsv, lower_belimbing, upper_belimbing)
mask_daun = cv2.inRange(hsv, lower_green, upper_green)
```

```python
segmentasi_belimbing = cv2.bitwise_and(image, image, mask=mask_belimbing)
segmentasi_daun = cv2.bitwise_and(image, image, mask=mask_daun)
```

```python
images = [image, mask_belimbing, segmentasi_belimbing, segmentasi_daun]
titles = ['Gambar Asli', 'Mask Belimbing', 'Segmentasi Belimbing', 'Segmentasi Daun']

plot_images(images, titles)
```
